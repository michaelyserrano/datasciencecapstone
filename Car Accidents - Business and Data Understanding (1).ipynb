# Applied Data Science Capstone - Car Accidents
### From IBM on Coursera

## Introduction / Business Problem

Motor accidents are often seen as unfortunate but inevitable daily occurances, costing money and lives. It would be invaluable for the Department of Transportation to be able to predict when and where accidents are most likely to happen in order to prepare both warning signs and medical personnel on days of high risk. While motor accidents are prevalent along every roadway, we will focus specifically on accidents in Somerville MA for this project.

## Data Understanding
A clustering model could be implemented to identify characteristics of historical accidents. In this case, a logistic regression model trained with traffic accident information from Somerville MA will cluster each different type of accident based on features such as weather, road conditions, the intersection and traffic way information, and the collision severity. For future accidents, information of weather and road conditions can be used to predict if a severe accident will occur on different areas of the road.
